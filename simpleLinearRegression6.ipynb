```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.rand(100, 1)
print(x)
```

```python
print(y)
```

```python
plt.plot(x, y, 'bo')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 2)
plt.ylim(0, 15)
plt.show()
```

```python
x_p = np.c_[np.ones((100, 1)), x]

print(x_p.shape)  # This will output (100, 2)
```

```python

print(x_p)
```

```python
theta_best = np.linalg.inv(x_p.T.dot(x_p)).dot(x_p.T).dot(y)

print(theta_best)  # This will output the best theta values
```

```python

x_test = np.array([[0], [2]])
x_test_p = np.c_[np.ones((2, 1)), x_test]

print(x_test)  # This will output the test input
```

```python
print(x_test_p)  # This will output the test input with intercept
```

```python
y_predict = x_test_p.dot(theta_best)
print(y_predict)  # This will output the predicted values
```

```python
plt.plot(x_test, y_predict, 'r-', label='Regression line')
plt.plot(x, y, 'bo')
plt.legend()
plt.show()
```
